```python
data['count'] = data.groupby('INVOICELINETEXT')['JobCardDate'].transform('size')
```

```python
data = data.set_index('JobCardDate')
```

```python
data['Year'] = data.index.year
```

```python
data = data.loc['2018']
```

```python
data = data.drop_duplicates(subset=['INVOICELINETEXT', 'Year','count'], keep='first')
```

```python
data['service']= data['INVOICELINETEXT'].map(str)+' '+data['count'].map(str)+' '+data['Year'].map(str)
```

```python
data.head()
```

```python
data.drop(['INVOICELINETEXT','count','Year'],axis=1,inplace=True)
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data.service=le.fit_transform(data.service)
```

```python
data.info()
```

```python
data.head()
```

```python
from matplotlib import pyplot
data.plot()
pyplot.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
X = data.values
size = int(len(X) * 0.9)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast(steps=365)[0]
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
```
